# 02 — Exploratory analysis (Phase 3)

This notebook turns cleaned tables into **charts** and adds a **plain-English interpretation** for each block.

**Rule:** if you cannot explain a figure in one short paragraph, do not put it in the dashboard yet.


```python
from __future__ import annotations

from pathlib import Path
import sys

import plotly.io as pio

pio.templates.default = "plotly_white"


def project_root() -> Path:
    here = Path.cwd().resolve()
    for p in [here, *here.parents]:
        if (p / "data" / "processed").is_dir():
            return p
    raise RuntimeError("Run from repo root")

ROOT = project_root()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.analysis.geo_enrichment import enrich_arsenic_for_maps, enrich_mosquito_for_maps
from src.analysis.loaders import load_arsenic_cleaned, load_mosquito_cleaned
from src.analysis import arsenic_patterns as ap
from src.analysis import mosquito_patterns as mp
from src.utils.paths import MI_COUNTIES_GEOJSON
from src.visualization.folium_maps import build_arsenic_folium_map, build_mosquito_folium_map

arsenic = load_arsenic_cleaned()
mosquito = load_mosquito_cleaned()
arsenic_geo = enrich_arsenic_for_maps(arsenic)
mosquito_geo = enrich_mosquito_for_maps(mosquito)
```

## Arsenic — distribution


```python
fig = ap.figure_arsenic_histogram(arsenic)
fig.show()
print(ap.interpret_histogram(arsenic))
```

## Arsenic — geography (ZIP-linked counties)


```python
fig = ap.figure_top_cities(arsenic, n=10)
fig.show()
print(ap.interpret_top_cities(arsenic))

fig2 = ap.figure_county_elevated_rate(arsenic_geo)
fig2.show()
print(ap.interpret_county_rates(arsenic_geo))
```

## Arsenic — temporal testing activity


```python
fig = ap.figure_tests_per_year(arsenic)
fig.show()
print(ap.interpret_yearly_trend(arsenic))
```

## Arsenic — severity / “elevated” share


```python
print(f"Elevated (reference band): {ap.elevated_rate(arsenic):.2f}%")
print(ap.interpret_elevated_share(arsenic))
```

## Mosquito — species frequency


```python
fig = mp.figure_species_counts(mosquito)
fig.show()
print(mp.interpret_species(mosquito))
```

## Mosquito — seasonal trends


```python
fig = mp.figure_season_counts(mosquito)
fig.show()
print(mp.interpret_season(mosquito))
```

## Mosquito — detection trends


```python
fig = mp.figure_detection_by_month(mosquito)
fig.show()
print(mp.interpret_detection(mosquito))
```

## Mosquito — site-type comparisons


```python
fig = mp.figure_site_type_totals(mosquito)
fig.show()

fig2 = mp.figure_site_type_avg_catch(mosquito)
fig2.show()

print(mp.interpret_site_types(mosquito))
```
